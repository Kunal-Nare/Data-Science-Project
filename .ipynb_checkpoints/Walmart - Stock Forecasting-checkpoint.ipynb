### Train Test Split

```python
feature_col_names = ['Dept','Size','Store','Wk_nbr','IsHoliday','Month','Year']
predicted_class_names = ['Weekly_Sales']
x = Walmart_train[feature_col_names].values
y = Walmart_train[predicted_class_names].values
split_test_size = 0.30
x_train, x_test, y_train, y_test = tts(x,y, test_size = split_test_size, random_state = 42) 
```

### Normalizing Data

```python
scaler = MinMaxScaler()
y_train = scaler.fit_transform(y_train) # fit and transform in one step
y_test = scaler.fit_transform(y_test)
```

### Function for Model

```python
def modelfit (alg,x_train,y_train,x_test,y_test):
    alg.fit(x_train,y_train.ravel())
    predict_train = alg.predict(x_train)
    predict_test = alg.predict(x_test)
    # Root Mean Squared Error on Test data
    cv_score = cross_val_score(alg, x_test,y_test , cv=5, scoring='neg_mean_squared_error')
    cv_score = np.sqrt(np.abs(cv_score))
    print("RMSE Test: %.4f" % np.sqrt(metrics.mean_squared_error(y_test, predict_test)))
    print("CV Score : Mean - %.4f | Std - %.4f | Min - %.4f | Max - %.4f" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
    # Mean Absolute Error on Test data
    cv_score = cross_val_score(alg, x_test,y_test , cv=5, scoring='neg_mean_absolute_error')
    cv_score = np.abs(cv_score)
    print("MAE Test: %.4f" % metrics.mean_absolute_error(y_test, predict_test))
    print("CV Score : Mean - %.4f | Std - %.4f | Min - %.4f | Max - %.4f" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
    print('R2 score',r2_score(y_test, predict_test))
```

### Algorthims

```python
LR = LinearRegression()
modelfit(LR,x_train,y_train,x_test,y_test)
```

```python
DT = DecisionTreeRegressor(random_state=0)
modelfit(DT,x_train,y_train,x_test,y_test)
```

```python
RF = RandomForestRegressor(random_state=0,n_estimators=100)
modelfit(RF,x_train,y_train.ravel(),x_test,y_test.ravel())
```

```python
GB = GradientBoostingRegressor()
modelfit(GB,x_train,y_train.ravel(),x_test,y_test.ravel())
```
